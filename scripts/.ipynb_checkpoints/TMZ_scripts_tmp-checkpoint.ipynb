```python
import pandas as pd
from scipy.stats import zscore
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import matplotlib
import numpy as np
from scipy import stats
import collections
import math
import os
import matplotlib as mpl


matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
def invitro_tmz_res_sen(df):
    """Use AUC and GR50 values from in vitro samples to separate TMZ sensitive and resistant samples.
    TMZ-sensitive if GR50 z-score <= 0 or AUC z-score <= 0
    TMZ-resistant if GR50 z-score > 0 or AUC z-score > 0"""
    
    #calculate GR50 zscore
    df.loc[df["GR50"] != float("inf"),"GR50_zscore"] = zscore(df.loc[df["GR50"] != float("inf"),"GR50"])
    
    #calculate AUC zscore for those that have GR50 value == inf
    df.loc[df["GR50"] == float("inf"),"AUC_zscore"] = zscore(df.loc[df["GR50"] == float("inf"),"AUC"])

    #separate Resistant and Sensitive
    df.loc[(df["GR50_zscore"] <= 0) | (df["AUC_zscore"] <= 0) , "in vitro Sensitivity"] = "S"
    df.loc[(df["GR50_zscore"] > 0) | (df["AUC_zscore"] > 0) , "in vitro Sensitivity"] = "R"

    #make a combined zscore column.
    df.loc[df["GR50"] != float("inf"),"combined zscore"] = df.loc[df["GR50"] != float("inf"),"GR50_zscore"]
    df.loc[df["GR50"] == float("inf"),"combined zscore"] = df.loc[df["GR50"] == float("inf"),"AUC_zscore"]

    return df
```

```python
def survivalcurve(dataframe, OSorPFS,OSPFS_yaxis, censor_column, group_column, 
                  group_1_name, group_2_name, group_1_value, group_2_value, unit, plot_1or0, savefilename):
    
    #group_1 will be red
    
    #%matplotlib inline
    import matplotlib as mpl
    label_size = 12
    mpl.rcParams['xtick.labelsize'] = label_size 
    mpl.rcParams['ytick.labelsize'] = label_size
    mpl.rcParams['figure.figsize'] = [4,3]
    #mpl.rcParams['font.family'] = 'Helvetica'
    


    import matplotlib.pyplot as plt
    import lifelines
    from lifelines import KaplanMeierFitter
    
    #plt.rc('font', family='Helvetica')
    
    kmf = KaplanMeierFitter()
    OScheck=dataframe.dropna(subset = [censor_column])

    T = OScheck[OSorPFS]
    C = OScheck[censor_column]

    ax = plt.subplot(111)

    group_1 = (OScheck[group_column] == group_1_value)##
    group_2 = (OScheck[group_column] == group_2_value)#

    kmf.fit(T[group_1], event_observed=C[group_1], label= group_1_name +" "+ str(sum(group_1)))
    kmf.plot(ax=ax, ci_show=False, show_censors=True, color = "#C00000",alpha =0.85, linewidth=2)
    kmf.fit(T[group_2], event_observed=C[group_2], label= group_2_name +" "+ str(sum(group_2)))
    kmf.plot(ax=ax, ci_show=False, show_censors=True, color = "#153FF5",alpha = 0.85, linewidth=2)

    from lifelines.statistics import logrank_test

    results = logrank_test(T[group_1], T[group_2], C[group_1], C[group_2], alpha=0.95 )

    plt.ylim(0,1);
    plt.legend(frameon=False, fontsize = label_size, loc=1)
    
    print("{0:.6f}".format(results.p_value))
    plt.xlabel(unit,fontsize = label_size)
    plt.ylabel(OSPFS_yaxis,fontsize = label_size)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    
    if plot_1or0 == 1:
        plt.savefig(savefilename,bbox_inches="tight")
        print("plt saved to: ",savefilename)
        plt.show()
    elif plot_1or0 == 0:
        plt.show()
```

```python
def tcga_4gene_survival(TCGAzscore, TCGAclinical):
    """Figure 3b"""
    
    survival = TCGAclinical.dropna(subset=["RNAseqID"])
    survival = survival.loc[(survival["IDH status"] == "WT")
                        &(survival["Sec.or.Rec"] == "No")
                        &(survival["TMZ(yes=1)"] == 1),:
                ]
    survival = survival.set_index("RNAseqID")


    #separate TMZ-resistant samples by TMZ-resistant marker gene expression
    TR_marker = ["PAPPA","ANXA3","LRRC3","EGR4"]

    TCGAzscore_1 = TCGAzscore.loc[TCGAzscore['Gene'].isin(TR_marker)] #zscore only from the gene list
    TCGAzscore_i = TCGAzscore_1[TCGAzscore_1.columns.intersection(survival["ID"])] #sample only from survival "Name" list
    TCGAzscore_i.insert(0, "Gene", TCGAzscore_1["Gene"])
    TCGAzscore_i = TCGAzscore_i.set_index("Gene")
    TCGAzscore_i = TCGAzscore_i.transpose()
    
    
    # select the samples OVER_expressed by 1 and 0
    threshold = 2

    TCGAzscore_i["altered"] = 0
    TCGAzscore_i.loc[(TCGAzscore_i > threshold).any(axis=1),"altered"] = 1
    TCGAzscore_i

    # merge with clinical data
    tcga = pd.merge(TCGAzscore_i, survival, how= "inner", left_on= TCGAzscore_i.index, right_on = "ID")

    #plot
    group = "altered"

    survivalcurve(tcga,"Disease Free (Months)","Progression Free Survival",'Disease Free Status', group,"High-risk","others",1,0,"months",
              1,"Figure3b-PFS.pdf")

    survivalcurve(tcga,"Overall Survival (Months)","Overall Survival",'Overall Survival Status', group,"High-risk","others",1,0,"months",
              1,"Figure3b-OS.pdf")
```

```python
def calc_RPKM(rc, genelength_column): 
    """calculates readcounts from featureCounts to RPKM.
    the index of rc needs to be gene names, and sample names in the columns.
    One of the column needs to be the length of the gene (genelength_column) to calculate RPKM"""

    dic={}

    for sample in rc.columns:
        dic[sample] = rc[sample].sum(axis=0)/1e6 # "SUM/1e6"

    for gene in rc.index:
        dic[gene] = rc.loc[gene, genelength_column]/1000 # "Length/1000"

    RPKM_c = pd.DataFrame(index = rc.index, columns = [x for x in rc.columns if genelength_column not in x])

    def divide_sum(x,sample): #divide by "SUM/1e6"
        return x/dic[sample]

    def divide_len(x,gene): # divide by "Length/1000"
        return x/dic[gene]


    for sample in RPKM_c.columns:
        RPKM_c[sample] = divide_sum(rc[sample],sample)

    RPKM_t = RPKM_c.transpose()

    for gene in RPKM_t.columns:
        RPKM_t[gene] = divide_len(RPKM_t[gene],gene)

    RPKM = RPKM_t.transpose()

    return RPKM
```

```python
def ir_compare_4gene(pair_fc):
    """Figure 3c"""
    pair_fc = pair_fc.set_index("Geneid")
    
    #calculate RPKM
    RPKM = calc_RPKM(pair_fc, "Length_x")

    #log2 transform
    lg2RPKM = np.log2(RPKM.iloc[:,:]+1)

    #quantilenorm
    rank_mean = lg2RPKM.stack().groupby(lg2RPKM.rank(method='first').stack().astype(int)).mean()
    lg2RPKM_qnorm = lg2RPKM.rank(method='min').stack().astype(int).map(rank_mean).unstack()

    #get the TMZ-resistant markers
    TR_marker = ["PAPPA","ANXA3","LRRC3","EGR4"]
    lg2RPKM_qnorm = lg2RPKM_qnorm.loc[TR_marker,:]


    #To make swarm plots, rearrange data for each gene, and plot
    
    for g in TR_marker:
        all_initials = lg2RPKM_qnorm.loc[g, [x for x in lg2RPKM_qnorm.columns if "RNA.1" in x]].tolist()
        all_recurrents = lg2RPKM_qnorm.loc[g, [x for x in lg2RPKM_qnorm.columns if "RNA.2" in x]].tolist()
        df= pd.DataFrame(index = ["IR","values"], data = [["I"]*40+["R"]*40,all_initials+all_recurrents])
        df = df.transpose()
        df["values"] = df["values"].astype('float')
        
        #calculate statistics
        stat, pval = stats.ranksums(all_initials,all_recurrents)
        print(g, stat, pval)
        
        #plot swarm plots
        fig = plt.figure(1, figsize=(3,4))

        #set color
        cls = sns.color_palette("Set2")
        my_pal = {"I": cls[0], "R": cls[1]}

        ax = sns.boxplot( x=df["IR"], y=df["values"], palette=my_pal, saturation=1, showfliers = False)
        ax = sns.swarmplot( x=df["IR"], y=df["values"], palette=my_pal, size=3, edgecolor='gray', linewidth=0.4)
        
        for patch in ax.artists:
            r, gr, b, a = patch.get_facecolor()
            patch.set_facecolor((r, gr, b, .3))
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)

        plt.ylabel("qnorm log2(RPKM+1)")

        plt.savefig("Figure3C_pair_%s.pdf"%g, bbox_inches = "tight")
        plt.show()
        
```

```python
def get_RS_numbers(df,f,x):
    """get the number of TMZ-Resistant/sensitive samples with type x in feature f.
    Calculate ratio, foldchange, fisher pvalue"""

    numR = len(df.loc[(df[f] == x)&(df["in vitro Sensitivity"] == "R"),:])
    numS = len(df.loc[(df[f] == x)&(df["in vitro Sensitivity"] == "S"),:])

    totR = len(df.loc[(~df[f].isnull())&(df[f] != "NA")&(df["in vitro Sensitivity"] == "R"),:])
    totS = len(df.loc[(~df[f].isnull())&(df[f] != "NA")&(df["in vitro Sensitivity"] == "S"),:])

    Rratio = numR/totR
    Sratio = numS/totS

    fc = (Rratio+1)/(Sratio+1)
    lg2fc = math.log(fc,2)

    #fisher exact
    oddsratio, pvalue = stats.fisher_exact([[numR,totR-numR],[numS,totS-numS]])

    return numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue
```

```python
def prep_bubbleplot(snv, cnv, metadata):
    """Figure 4a"""
    
    #get the total number of samples in the main cohort that has WES and gliomaSCAN for SNV detection
    main_snv_S = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") 
                              &(~metadata["DNAseq availability"].isnull())
                              &(metadata["in vitro Sensitivity"] == "S"),"sample ID"].tolist()
    
    main_snv_R = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") 
                              &(~metadata["DNAseq availability"].isnull())
                              &(metadata["in vitro Sensitivity"] == "R"),"sample ID"].tolist()
    
    #get the total number of samples in the main cohort that has WES , gliomaSCAN, and RNAseq for CNV detection
    main_cnv_S = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") 
                                  &((~metadata["DNAseq availability"].isnull()) | (~metadata["Tissue RNAseq"].isnull()))
                                  &(metadata["in vitro Sensitivity"] == "S"),"sample ID"].tolist()
    
    main_cnv_R = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") 
                                  &((~metadata["DNAseq availability"].isnull()) | (~metadata["Tissue RNAseq"].isnull()))
                                  &(metadata["in vitro Sensitivity"] == "R"),"sample ID"].tolist()
    
    
    #SNV: merge SAVI2(WES) and GliomaSCAN results
    major_genes = ["PTEN","PIK3R1","PIK3CA","PIK3CG","PDGFRA","EGFR","NF1","NF2","RB1","TP53","ATRX"]
    snv = snv.loc[snv["Gene_Name"].isin(major_genes),:]
    
    snv["uniq_mutations"] = snv.apply(lambda x: x["ID"] + "_" + str(x["position"]),  axis=1)
    snv.drop_duplicates("uniq_mutations", inplace = True)
    
    #insert in vitro sensitivity
    snv = pd.merge(metadata[["sample ID","Main cohort (n = 69)","in vitro Sensitivity"]], snv, how = "inner", left_on = "sample ID", right_on = "ID")
    snv
    
    #get only main cohort
    snv = snv.loc[snv["Main cohort (n = 69)"] == "yes",:]
    
    #Compare mutations in Resistant and Sensitive sample
    mut = []
    for gene in major_genes: 
        samples = list(set(snv.loc[snv["Gene_Name"] == gene,"ID"]))
        rs=[]
        for s in samples:
            rs.append(snv.loc[snv["ID"] == s,"in vitro Sensitivity"].tolist()[0]) 
        mut.append([gene, rs.count("R"), rs.count("S"),len(main_snv_R),len(main_snv_S)])
    
    
    #CNV
    cnv = cnv.set_index("ID")
    
    cnv_res=[]
    ampdel=['PTEN_del','CDKN2A/B_del','PDGFRA_amp','CDK4_amp','MDM2_amp','EGFR_amp']
    
    ########USE ONLY DEEP DEL/ AMP for BUBBLE PLOT
    for i in [x for x in ampdel if "amp" in x]:
        
        gene = i.replace('_amp', '')
        
        tot = cnv.loc[cnv[gene]=="amp",:].index.tolist()

        altered_r = len([ x for x in tot if x in main_cnv_R])
        altered_s = len([ x for x in tot if x in main_cnv_S])

        cnv_res.append([i,altered_r,altered_s,len(main_cnv_R),len(main_cnv_S)])

    for i in [x for x in ampdel if "del" in x]:
        
        gene = i.replace('_del', '')
        
        tot = cnv.loc[cnv[gene]=="del",:].index.tolist()

        altered_r = len([ x for x in tot if x in main_cnv_R])
        altered_s = len([ x for x in tot if x in main_cnv_S])

        cnv_res.append([i,altered_r,altered_s,len(main_cnv_R),len(main_cnv_S)])

        
    #combine mut and cnv

    al = cnv_res + mut
    al = pd.DataFrame(al, columns=["gene","R","S","totR","totS"]) 

    al["R/totR"] = al["R"]/al["totR"]
    al["S/totS"] = al["S"]/al["totS"]
    al["fc(r_ratio+1/s_ratio+1)"] = (al["R/totR"]+1)/(al["S/totS"]+1)

    for i in al.index:
        al.loc[i, "log2FC"] = math.log(al.loc[i,"fc(r_ratio+1/s_ratio+1)"],2)
        oddsratio, pvalue = stats.fisher_exact([
            [al.loc[i, "R"], al.loc[i, "totR"]-al.loc[i, "R"]], 
            [al.loc[i, "S"], al.loc[i, "totS"]-al.loc[i, "S"]]])

        al.loc[i,"fisher_P"] = pvalue
    al = al.sort_values(by="log2FC")#,inplace=True)
    al = al.reset_index(drop=True)

    #insert MGMT promoter status
    metamain = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes"),:]
    
        
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"MGMT promoter methylation status","M")
    MGMTm = ["MGMT methylated",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]
    
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"MGMT promoter methylation status","UM")
    MGMTum = ["MGMT unmethylated",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]

    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"hypermutation","hypermutation")
    hypermut = ["hypermutation", numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]

    #5-ala calc on main+multisector samples
    metamainmulti = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") | 
                                 (metadata["Multi-sector cohort"] == "yes"),:]
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamainmulti,"5-ALA","Positive")
    ala5p = ["5-ALA Positive",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamainmulti,"5-ALA","Negative")
    ala5n = ["5-ALA Negative",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]

    #GBM subtypes
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"Tissue GBM subtype","Classical")
    subtype_classical = ["subtype_classical",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"Tissue GBM subtype","Mesenchymal")
    subtype_mesenchymal = ["subtype_mesenchymal",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"Tissue GBM subtype","Proneural")
    subtype_proneural = ["subtype_proneural",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]

    #TMZ resistant markers (4 gene) #from RPKM lg2 qnorm values, ttest_ind
    MGMTexp = ["MGMT_exp","-","-",12,22,"-","-",1.246,0.318,0.0135]
    EGR4exp = ["EGR4_exp","-","-",12,22,"-","-",1.989,0.992,0.00858]
    PAPPAexp = ["PAPPA_exp","-","-",12,22,"-","-",1.452,0.538,0.00516]
    LRRC3exp = ["LRRC3_exp","-","-",12,22,"-","-",1.324,0.405,0.0218]
    ANXA3exp = ["ANXA3_exp","-","-",12,22,"-","-",1.532,0.615,0.000696]

    #Add to dataframe
    al.loc[len(al),:] = MGMTm
    al.loc[len(al),:] = MGMTum
    al.loc[len(al),:] = hypermut
    al.loc[len(al),:] = ala5n
    al.loc[len(al),:] = ala5p
    al.loc[len(al),:] = subtype_classical
    al.loc[len(al),:] = subtype_mesenchymal
    al.loc[len(al),:] = subtype_proneural
    al.loc[len(al),:] = MGMTexp
    al.loc[len(al),:] = EGR4exp
    al.loc[len(al),:] = PAPPAexp
    al.loc[len(al),:] = LRRC3exp
    al.loc[len(al),:] = ANXA3exp

    al = al.sort_values(by="log2FC")
    al = al.reset_index(drop=True)
    
    return al
```

```python
def plot_bubbleplot(al):
    %matplotlib inline

    plt.style.use('seaborn-white')
    label_size = 14
    mpl.rcParams['xtick.labelsize'] = label_size 
    mpl.rcParams['ytick.labelsize'] = label_size
    mpl.rcParams['figure.figsize'] = [12,17]

    x = al["log2FC"].tolist()
    y = al.index.tolist()

    fig, ax = plt.subplots()
    s=list(range(len(al)))
    s = [x*50 for x in s]

    #add grey line
    ax.axvline(0, linestyle='--', color='grey',alpha=0.8)

    #remove spines
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)

    msize=[]
    for i in al["fisher_P"]:
        if i >= 0.5:
            sss = (math.log(1/(1),10))*600 +10
            msize.append(sss)
        else:
            sss = (math.log(1/(i),10))*600 +10
            msize.append(sss)




    #count number of red/blue
    b = len(al.loc[al["log2FC"] < 0,:])
    r = len(al.loc[al["log2FC"] > 0,:])

    #add text
    t_size = 14
    t_color = "black"

    for i,txt,pval in zip(range(len(al["gene"][:b])),
                          al["gene"][:b].tolist(),
                         al["fisher_P"][:b].tolist()):

        if pval >= 0.5:
            j = -8
            down = -4
        elif (pval < 0.5) and (pval > 0.1):
            j = -12
            down = -4
        elif pval < 0.1:
            j = -20
            down = -3


        #j = -(msize[i]/60)-12
        if ("subtype" not in txt ) and ("_" in txt):
            txtl = txt.split("_")
            ax.annotate(r'$%s$_%s'%(txtl[0],txtl[1]), (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
        elif ("ALA" not in txt ) and  (" " in txt):
            txtl = txt.split(" ")    
            ax.annotate(r'$%s$ %s'%(txtl[0],txtl[1]), (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
        elif "ALA" in txt:
            ax.annotate(txt, (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
        elif "subtype" in txt:
            ax.annotate(txt, (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
        else:
            ax.annotate(r'$%s$'%txt, (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
    #for i, txt,pval in enumerate(al["gene"][b:b+r].tolist()):
    for i,txt,pval in zip(range(len(al["gene"][b:b+r])),
                          al["gene"][b:b+r].tolist(),
                         al["fisher_P"][b:b+r].tolist()):

        #j = (msize[i+b]/60)+10
        if pval >= 0.5:
            j = 8
            down = -4.5
        elif (pval < 0.5) and (pval > 0.1):
            j = 10
            down = -6
        elif (pval < 0.1) and (pval > 0.01):
            j = 20
            down = -8
        elif (pval < 0.01):
            j = 22
            down = -8

        if ("subtype" not in txt ) and ("_" in txt):
            txtl = txt.split("_")
            ax.annotate(r'$%s$_%s'%(txtl[0],txtl[1]), (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)
        elif ("ALA" not in txt ) and  (" " in txt):
            txtl = txt.split(" ")    
            ax.annotate(r'$%s$ %s'%(txtl[0],txtl[1]), (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)

        elif "hypermutation" in txt:
            ax.annotate(txt, (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)
        elif "ALA" in txt:
            ax.annotate(txt, (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)
        elif "subtype" in txt:
            ax.annotate(txt, (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)
        else:
            ax.annotate(r'$%s$'%txt, (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)


    plt.rc('font', family='Helvetica')
    plt.xticks(size=label_size)
    plt.scatter(x,y,s=msize,color = ["#153FF5"]*b+["#C00000"]*r, alpha =0.5, marker='o')
    ax.get_yaxis().set_visible(False)
    plt.xlim([-0.6, 1.2])
    plt.savefig("Figure4a.pdf", bbox_inches="tight")
    plt.show()

    #add legend
    
    fig, ax = plt.subplots()
    s=list(range(len(al)))
    s = [x*50 for x in s]

    #remove spines
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)

    #set marker size
    msize = [(math.log(1/(x),10))*600 +10 for x in [0.001, 0.01, 0.1, 1]] #math.log(1/(x),10))*1200 +50

    ax.annotate(">0.5", (0.405, 0.3))
    ax.annotate("0.1", (0.405, 0.25))
    ax.annotate("0.01", (0.405, 0.18))
    ax.annotate("0.001", (0.405, 0.09))
    plt.scatter([0.4]*4,[0.1,0.18,0.25,0.3],s=msize,color ="grey", alpha =0.5, marker='o')
    plt.ylim(0,1.5)
    plt.savefig("Figure4a_legend.pdf", bbox_inches="tight")
    plt.show()
```

```python
def prep_bubbleplot(snv, cnv, metadata):
    """Figure 4a"""
    
    #get the total number of samples in the main cohort that has WES and gliomaSCAN for SNV detection
    main_snv_S = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") 
                              &(~metadata["DNAseq availability"].isnull())
                              &(metadata["in vitro Sensitivity"] == "S"),"sample ID"].tolist()
    
    main_snv_R = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") 
                              &(~metadata["DNAseq availability"].isnull())
                              &(metadata["in vitro Sensitivity"] == "R"),"sample ID"].tolist()
    
    #get the total number of samples in the main cohort that has WES , gliomaSCAN, and RNAseq for CNV detection
    main_cnv_S = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") 
                                  &((~metadata["DNAseq availability"].isnull()) | (~metadata["Tissue RNAseq"].isnull()))
                                  &(metadata["in vitro Sensitivity"] == "S"),"sample ID"].tolist()
    
    main_cnv_R = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") 
                                  &((~metadata["DNAseq availability"].isnull()) | (~metadata["Tissue RNAseq"].isnull()))
                                  &(metadata["in vitro Sensitivity"] == "R"),"sample ID"].tolist()
    
    
    #SNV: merge SAVI2(WES) and GliomaSCAN results
    major_genes = ["PTEN","PIK3R1","PIK3CA","PIK3CG","PDGFRA","EGFR","NF1","NF2","RB1","TP53","ATRX"]
    snv = snv.loc[snv["Gene_Name"].isin(major_genes),:]
    
    snv["uniq_mutations"] = snv.apply(lambda x: x["ID"] + "_" + str(x["position"]),  axis=1)
    snv.drop_duplicates("uniq_mutations", inplace = True)
    
    #insert in vitro sensitivity
    snv = pd.merge(metadata[["sample ID","Main cohort (n = 69)","in vitro Sensitivity"]], snv, how = "inner", left_on = "sample ID", right_on = "ID")
    snv
    
    #get only main cohort
    snv = snv.loc[snv["Main cohort (n = 69)"] == "yes",:]
    
    #Compare mutations in Resistant and Sensitive sample
    mut = []
    for gene in major_genes: 
        samples = list(set(snv.loc[snv["Gene_Name"] == gene,"ID"]))
        rs=[]
        for s in samples:
            rs.append(snv.loc[snv["ID"] == s,"in vitro Sensitivity"].tolist()[0]) 
        mut.append([gene, rs.count("R"), rs.count("S"),len(main_snv_R),len(main_snv_S)])
    
    
    #CNV
    cnv = cnv.set_index("ID")
    
    cnv_res=[]
    ampdel=['PTEN_del','CDKN2A/B_del','PDGFRA_amp','CDK4_amp','MDM2_amp','EGFR_amp']
    
    ########USE ONLY DEEP DEL/ AMP for BUBBLE PLOT
    for i in [x for x in ampdel if "amp" in x]:
        
        gene = i.replace('_amp', '')
        
        tot = cnv.loc[cnv[gene]=="amp",:].index.tolist()

        altered_r = len([ x for x in tot if x in main_cnv_R])
        altered_s = len([ x for x in tot if x in main_cnv_S])

        cnv_res.append([i,altered_r,altered_s,len(main_cnv_R),len(main_cnv_S)])

    for i in [x for x in ampdel if "del" in x]:
        
        gene = i.replace('_del', '')
        
        tot = cnv.loc[cnv[gene]=="del",:].index.tolist()

        altered_r = len([ x for x in tot if x in main_cnv_R])
        altered_s = len([ x for x in tot if x in main_cnv_S])

        cnv_res.append([i,altered_r,altered_s,len(main_cnv_R),len(main_cnv_S)])

        
    #combine mut and cnv

    al = cnv_res + mut
    al = pd.DataFrame(al, columns=["gene","R","S","totR","totS"]) 

    al["R/totR"] = al["R"]/al["totR"]
    al["S/totS"] = al["S"]/al["totS"]
    al["fc(r_ratio+1/s_ratio+1)"] = (al["R/totR"]+1)/(al["S/totS"]+1)

    for i in al.index:
        al.loc[i, "log2FC"] = math.log(al.loc[i,"fc(r_ratio+1/s_ratio+1)"],2)
        oddsratio, pvalue = stats.fisher_exact([
            [al.loc[i, "R"], al.loc[i, "totR"]-al.loc[i, "R"]], 
            [al.loc[i, "S"], al.loc[i, "totS"]-al.loc[i, "S"]]])

        al.loc[i,"fisher_P"] = pvalue
    al = al.sort_values(by="log2FC")#,inplace=True)
    al = al.reset_index(drop=True)

    #insert MGMT promoter status
    metamain = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes"),:]
    
    def get_RS_numbers(df,f,x):
        """get the number of TMZ-Resistant/sensitive samples with type x in feature f.
        Calculate ratio, foldchange, fisher pvalue"""
        
        numR = len(df.loc[(df[f] == x)&(df["in vitro Sensitivity"] == "R"),:])
        numS = len(df.loc[(df[f] == x)&(df["in vitro Sensitivity"] == "S"),:])
        
        totR = len(df.loc[(~df[f].isnull())&(df[f] != "NA")&(df["in vitro Sensitivity"] == "R"),:])
        totS = len(df.loc[(~df[f].isnull())&(df[f] != "NA")&(df["in vitro Sensitivity"] == "S"),:])
        
        Rratio = numR/totR
        Sratio = numS/totS
        
        fc = (Rratio+1)/(Sratio+1)
        lg2fc = math.log(fc,2)
        
        #fisher exact
        oddsratio, pvalue = stats.fisher_exact([[numR,totR-numR],[numS,totS-numS]])
        
        return numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue
    
        
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"MGMT promoter methylation status","M")
    MGMTm = ["MGMT methylated",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]
    
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"MGMT promoter methylation status","UM")
    MGMTum = ["MGMT unmethylated",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]

    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"hypermutation","hypermutation")
    hypermut = ["hypermutation", numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]

    #5-ala calc on main+multisector samples
    metamainmulti = metadata.loc[(metadata["Main cohort (n = 69)"] == "yes") | 
                                 (metadata["Multi-sector cohort"] == "yes"),:]
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamainmulti,"5-ALA","Positive")
    ala5p = ["5-ALA Positive",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamainmulti,"5-ALA","Negative")
    ala5n = ["5-ALA Negative",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]

    #GBM subtypes
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"Tissue GBM subtype","Classical")
    subtype_classical = ["subtype_classical",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"Tissue GBM subtype","Mesenchymal")
    subtype_mesenchymal = ["subtype_mesenchymal",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]
    numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue = get_RS_numbers(metamain,"Tissue GBM subtype","Proneural")
    subtype_proneural = ["subtype_proneural",numR,numS,totR,totS,Rratio,Sratio,fc,lg2fc,pvalue]

    #TMZ resistant markers (4 gene) #from RPKM lg2 qnorm values, ttest_ind
    MGMTexp = ["MGMT_exp","-","-",12,22,"-","-",1.246,0.318,0.0135]
    EGR4exp = ["EGR4_exp","-","-",12,22,"-","-",1.989,0.992,0.00858]
    PAPPAexp = ["PAPPA_exp","-","-",12,22,"-","-",1.452,0.538,0.00516]
    LRRC3exp = ["LRRC3_exp","-","-",12,22,"-","-",1.324,0.405,0.0218]
    ANXA3exp = ["ANXA3_exp","-","-",12,22,"-","-",1.532,0.615,0.000696]

    #Add to dataframe
    al.loc[len(al),:] = MGMTm
    al.loc[len(al),:] = MGMTum
    al.loc[len(al),:] = hypermut
    al.loc[len(al),:] = ala5n
    al.loc[len(al),:] = ala5p
    al.loc[len(al),:] = subtype_classical
    al.loc[len(al),:] = subtype_mesenchymal
    al.loc[len(al),:] = subtype_proneural
    al.loc[len(al),:] = MGMTexp
    al.loc[len(al),:] = EGR4exp
    al.loc[len(al),:] = PAPPAexp
    al.loc[len(al),:] = LRRC3exp
    al.loc[len(al),:] = ANXA3exp

    al = al.sort_values(by="log2FC")
    al = al.reset_index(drop=True)
    
    return al


def plot_bubbleplot(al):
    %matplotlib inline

    plt.style.use('seaborn-white')
    label_size = 14
    mpl.rcParams['xtick.labelsize'] = label_size 
    mpl.rcParams['ytick.labelsize'] = label_size
    mpl.rcParams['figure.figsize'] = [12,17]

    x = al["log2FC"].tolist()
    y = al.index.tolist()

    fig, ax = plt.subplots()
    s=list(range(len(al)))
    s = [x*50 for x in s]

    #add grey line
    ax.axvline(0, linestyle='--', color='grey',alpha=0.8)

    #remove spines
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)

    msize=[]
    for i in al["fisher_P"]:
        if i >= 0.5:
            sss = (math.log(1/(1),10))*600 +10
            msize.append(sss)
        else:
            sss = (math.log(1/(i),10))*600 +10
            msize.append(sss)




    #count number of red/blue
    b = len(al.loc[al["log2FC"] < 0,:])
    r = len(al.loc[al["log2FC"] > 0,:])

    #add text
    t_size = 14
    t_color = "black"

    for i,txt,pval in zip(range(len(al["gene"][:b])),
                          al["gene"][:b].tolist(),
                         al["fisher_P"][:b].tolist()):

        if pval >= 0.5:
            j = -8
            down = -4
        elif (pval < 0.5) and (pval > 0.1):
            j = -12
            down = -4
        elif pval < 0.1:
            j = -20
            down = -3


        #j = -(msize[i]/60)-12
        if ("subtype" not in txt ) and ("_" in txt):
            txtl = txt.split("_")
            ax.annotate(r'$%s$_%s'%(txtl[0],txtl[1]), (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
        elif ("ALA" not in txt ) and  (" " in txt):
            txtl = txt.split(" ")    
            ax.annotate(r'$%s$ %s'%(txtl[0],txtl[1]), (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
        elif "ALA" in txt:
            ax.annotate(txt, (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
        elif "subtype" in txt:
            ax.annotate(txt, (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
        else:
            ax.annotate(r'$%s$'%txt, (x[i], y[i]),textcoords="offset points",ha='right',
                    xytext=(-(-x[i]+1)+j,down ),#fontweight='bold',
                    size=t_size, color =t_color)
    #for i, txt,pval in enumerate(al["gene"][b:b+r].tolist()):
    for i,txt,pval in zip(range(len(al["gene"][b:b+r])),
                          al["gene"][b:b+r].tolist(),
                         al["fisher_P"][b:b+r].tolist()):

        #j = (msize[i+b]/60)+10
        if pval >= 0.5:
            j = 8
            down = -4.5
        elif (pval < 0.5) and (pval > 0.1):
            j = 10
            down = -6
        elif (pval < 0.1) and (pval > 0.01):
            j = 20
            down = -8
        elif (pval < 0.01):
            j = 22
            down = -8

        if ("subtype" not in txt ) and ("_" in txt):
            txtl = txt.split("_")
            ax.annotate(r'$%s$_%s'%(txtl[0],txtl[1]), (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)
        elif ("ALA" not in txt ) and  (" " in txt):
            txtl = txt.split(" ")    
            ax.annotate(r'$%s$ %s'%(txtl[0],txtl[1]), (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)

        elif "hypermutation" in txt:
            ax.annotate(txt, (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)
        elif "ALA" in txt:
            ax.annotate(txt, (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)
        elif "subtype" in txt:
            ax.annotate(txt, (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)
        else:
            ax.annotate(r'$%s$'%txt, (x[i+b], y[i+b]),textcoords="offset points",ha='left',
                    xytext=((x[i]+1)+j,down),
                    size=t_size, color =t_color)


    plt.rc('font', family='Helvetica')
    plt.xticks(size=label_size)
    plt.scatter(x,y,s=msize,color = ["#153FF5"]*b+["#C00000"]*r, alpha =0.5, marker='o')
    ax.get_yaxis().set_visible(False)
    plt.xlim([-0.6, 1.2])
    plt.savefig("Figure4a.pdf", bbox_inches="tight")
    plt.show()

    #add legend
    
    fig, ax = plt.subplots()
    s=list(range(len(al)))
    s = [x*50 for x in s]

    #remove spines
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)

    #set marker size
    msize = [(math.log(1/(x),10))*600 +10 for x in [0.001, 0.01, 0.1, 1]] #math.log(1/(x),10))*1200 +50

    ax.annotate(">0.5", (0.405, 0.3))
    ax.annotate("0.1", (0.405, 0.25))
    ax.annotate("0.01", (0.405, 0.18))
    ax.annotate("0.001", (0.405, 0.09))
    plt.scatter([0.4]*4,[0.1,0.18,0.25,0.3],s=msize,color ="grey", alpha =0.5, marker='o')
    plt.ylim(0,1.5)
    plt.savefig("Figure4a_legend.pdf", bbox_inches="tight")
    plt.show()

```
